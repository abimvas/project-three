```python
import pandas as pd
import os
import dateutil
import datetime
import csv
```

```python
df = df.replace({"Côte D'Ivoire": u"C\xf4te d'Ivoire"})
df = df.replace({"Syria": "Syrian Arab Republic"})
df = df.replace({"Tanzania": "Tanzania, United Republic of"})
df = df.replace({"Vietnam": "Viet Nam"})
df = df.replace({"Congo (Democratic Republic Of The)": "Congo, The Democratic Republic of the"})
df = df.replace({"Iran": "Iran, Islamic Republic of"})
df = df.replace({"Russia": "Russian Federation"})       
df = df.replace({"Burma": "Myanmar"})
df = df.replace({"South Korea": "Korea, Republic of"})
df = df.replace({"Taiwan": "Taiwan, Province of China"})
```

```python
df.head()
```

```python
df['City'].unique()
```

```python
df = df.replace({"Bogotá": "Bogata"})
df = df.replace({"São Paulo": "Sao Paulo"})
```

```python
df['City'].unique()
```

```python
df.head()
```

```python
import pycountry
```

```python
country_codes = {}
for country in pycountry.countries:
    country_codes[country.name] = country.alpha_2

codes = [country_codes.get(country) for country in df["Country"]]

print (codes)  # prints [u'AS', u'CA', u'FR']
```

```python
df["CountryCode"] = codes
```

```python
df.head()
```

```python
df.dtypes
```

```python
lat_dir = df.Latitude.str.split('(\d+\.\d+)(N|S|E|W)', expand=True)
lng_dir = df.Longitude.str.split('(\d+\.\d+)(N|S|E|W)', expand=True)
```

```python
df.LatitudeNum = df.Latitude.str.split('(\d+\.\d+)(N|S|E|W)', expand=True).loc[:, 2]
```

```python
lat_dir
```

```python
lat_parts = df.Latitude.str.split(('(\d+\.\d+)(N|S|E|W)'), expand=True)
lat_parts["new_lat"] = lat_parts[1].astype(float) * lat_parts[2].map({'N':1, 'S':-1, 'E': 1, 'W':-1});


lng_parts = df.Longitude.str.split(('(\d+\.\d+)(N|S|E|W)'), expand=True)
lng_parts["new_lng"] = lng_parts[1].astype(float) * lng_parts[2].map({'N':1, 'S':-1, 'E': 1, 'W':-1});
```

```python
lng_parts['new_lng']
```

```python
df.dt = pd.to_datetime(df.dt)

df['Year'] = df.dt.dt.year
df['Month'] = df.dt.dt.month
df['Day'] = df.dt.dt.day
```

```python
Lat_Dir = lat_parts[2]
df.insert(8,'Lat_Dir',Lat_Dir)

Lat_Coord = lat_parts['new_lat']
df.insert(7,'Lat_Coord',Lat_Coord)

Lng_Dir = lng_parts[2]
df.insert(8,'Lng_Dir',Lng_Dir)

Lng_Coord = lng_parts['new_lng']
df.insert(8,'Lng_Coord',Lng_Coord)
```

```python
df.head()
```

```python
df = df.rename(columns={"dt":"Date", "Year": "Yr", "Month": "M", "Day": "D", "AverageTemperature":"AvgTemp", 
                        "AverageTemperatureUncertainty":"AvgTUncert", "City":"City", "Country":"Country",
                        "CountryCode":"CountryCode","Latitude":"Lat", "Longitude":"Lon",
                        "Lat_Coord": "Lat_Coord", "Lat_Dir": "Lat_Dir", "Lng_Coord": "Lng_Coord", 
                        "Lng_Dir": "Lng_Dir"  
                  })
df.head()
```
