```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy import create_engine, MetaData
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Numeric, Text, Float
```

```python
# Engine for database `hurricanes.sqlite`
engine = create_engine("sqlite:///hurricanes.sqlite")
```

```python
# Create a connection to the engine called `conn`
conn = engine.connect()
```

```python
# Use `declarative_base` from SQLAlchemy to model the demographics table as an ORM class
# Make sure to specify types for each column
### BEGIN SOLUTION

Base = declarative_base()

class Hurricanes(Base):
    __tablename__ = 'hurricanes'

    id = Column(Integer, primary_key=True)
    Serial_Num = Column(Integer)
    Season_year = Column(Integer)
    Num = Column(Integer)
    Basin = Column(Text)
    Sub_basin = Column(Text)
    Name = Column(Text)
    ISO_time = Column(Text)
    Nature = Column(Text)
    Latitude = Column(Integer)
    Longitude = Column(Integer)
    Wind_WMO = Column(Integer)
    Pres_WMO = Column(Integer)
    Center = Column(Integer)
    Wind_Perc = Column(Integer)
    Pres_Perc = Column(Integer)

    def __repr__(self):
        return f"id={self.id}, name={self.name}"
### END SOLUTION
```

```python
# Use `create_all` to create the hurricanes table in the database
Base.metadata.create_all(engine)
```

```python
# Load the cleaned csv file into a pandas dataframe
new_df = pd.read_csv(new_csv)
```

```python
# Use Orient='records' to create a list of data to write
# http://pandas-docs.github.io/pandas-docs-travis/io.html#orient-options
### BEGIN SOLUTION
data = new_df.to_dict(orient='records')
data[0]
### END SOLUTION
```

```python
# Use MetaData from SQLAlchemy to reflect the tables
metadata = MetaData(bind=engine)
metadata.reflect()
```

```python
# Save the reference to the `hurricanes` table as a variable called `table`
table = sqlalchemy.Table('hurricanes', metadata, autoload=True)
```

```python
# Use `table.delete()` to remove any existing data.
# Only use when the entire code needs to re-run from creating sqlalchemy

conn.execute(table.delete())
```

```python
# Use `table.insert()` to insert the data into the table

conn.execute(table.insert(), data)
```

```python
# Test that the insert works by fetching the first 5 rows. 
conn.execute("select * from hurricanes limit 5").fetchall()
```
